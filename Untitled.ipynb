```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import boto3
import io

# 1. Configuration - Pointing to your specific 'Gold' layer file
bucket = 'cs480-injury-mortality-data'
key = 'injury_mortality_final_cleaned (1).csv'
s3 = boto3.client('s3')

# 2. Ingest Data from S3
response = s3.get_object(Bucket=bucket, Key=key)
df = pd.read_csv(io.BytesIO(response['Body'].read()))

# 3. Data Cleaning: Remove aggregate 'Summary' rows to prevent double-counting
# This is a key technical insight for your report.
df_clean = df[df['Is_Summary_Row'] == 0].copy()

# 4. Feature Engineering: Create the 'Risk Category' Calculated Field
# Logic: Target_Rate > 50 is High Priority for public health intervention.
df_clean['Risk_Category'] = df_clean['Target_Rate'].apply(
    lambda x: 'High Priority' if x > 50 else 'Standard'
)

# --- VISUAL 1: Historical Trend (The Line Chart) ---
plt.figure(figsize=(10, 5))
sns.set_style("whitegrid")
sns.lineplot(data=df_clean, x='Year', y='Target_Rate', color='#3498db', linewidth=2.5)
plt.title('20-Year Injury Mortality Rate Trend', fontsize=15)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Mortality Rate (per 100k)', fontsize=12)
plt.show() # RIGHT-CLICK AND SAVE AS 'Line_Chart_Trends.png'

# --- VISUAL 2: Lethality by Mechanism (The Purple Bar Chart) ---
plt.figure(figsize=(10, 8))
# Grouping by mechanism and calculating mean lethality
mech_data = df_clean.groupby('Injury mechanism')['Target_Rate'].mean().sort_values(ascending=False).head(10)
sns.barplot(x=mech_data.values, y=mech_data.index, palette='magma')
plt.title('Top 10 High-Lethality Injury Mechanisms', fontsize=15)
plt.xlabel('Mean Mortality Rate', fontsize=12)
plt.ylabel('Mechanism', fontsize=12)
plt.show() # RIGHT-CLICK AND SAVE AS 'Bar_Chart_Mechanisms.png'

# --- VISUAL 3: Risk Category Distribution (The Business Logic Chart) ---
plt.figure(figsize=(8, 5))
sns.countplot(data=df_clean, x='Risk_Category', palette=['#e74c3c', '#34495e'])
plt.title('Distribution of Records by Risk Category', fontsize=15)
plt.show() # RIGHT-CLICK AND SAVE AS 'Risk_Distribution.png'
```

```python
import sagemaker
import boto3
from sagemaker.inputs import TrainingInput
from sagemaker import image_uris

# 1. Setup Session
sess = sagemaker.Session()
role = sagemaker.get_execution_role()
bucket = 'cs480-injury-mortality-data'

# 2. Algorithm Choice (Rubric Requirement)
# Using SageMaker's built-in XGBoost container
container = image_uris.retrieve("xgboost", sess.boto_region_name, "1.5-1")

# 3. Define the Training Input (Reading from your S3 Gold/Training folders)
s3_input_train = TrainingInput(s3_data=f's3://{bucket}/training-input/train', content_type='csv')
s3_input_val = TrainingInput(s3_data=f's3://{bucket}/training-input/validation', content_type='csv')

# 4. The Estimator Configuration
xgb = sagemaker.estimator.Estimator(
    image_uri=container,
    role=role,
    instance_count=1,
    instance_type='ml.m5.large', # Cost-efficient compute
    output_path=f's3://{bucket}/training-output',
    sagemaker_session=sess
)

# 5. Hyperparameters (The "Tuning" logic for your slides)
xgb.set_hyperparameters(
    max_depth=5,            # Depth of the trees
    eta=0.2,                # Learning rate
    objective='reg:squarederror', 
    num_round=100           # Number of boosting iterations
)

# 6. Execute Training
# Note: You already ran this, so you can just capture the output logs
print("XGBoost Model Architecture defined.")
print(f"Algorithm: {container}")
print("Ready for Slide 4 documentation.")
```
